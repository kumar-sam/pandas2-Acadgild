```python
# Assignment 9
import pandas as pd  #importing pandas library
df=pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")
df.head() #printing the first 5 tuples of df
```

```python
#Assignment 9.1
"""
Delete unnamed columns
"""
del df['Unnamed: 0'] #deleting the column which is named 'Unnamed: 0'
df.head()
```

```python
#Assignment 9.2
"""
Show the distribution of male and female
"""
gender_categ=pd.Categorical(df['Gender']) #using categorical attribute to categorize them
gender_categ.describe() #describing the categorical
```

```python
#Assignment 9.3
"""
Show the top 5 most preferred names
"""
df_sorted=df.sort_values(['Count'],ascending=False) #sorting the dataframe into descending order 
df_sorted.head(5) #printing the top 5 tuples
```

```python
#Assignment 9.4
"""
What is the median name occurence in the dataset
"""
df_sorted.head(5)
no_of_tuples =df_sorted.shape[0] #no_of_tuples stores the number of tuples in data frame

if(no_of_tuples %2 ==0): #if n mod 2 ==0 ,then median is val[floor(n/2)]
    print("median of the data frame is ",df_sorted.iloc[int(no_of_tuples/2),1])
else: #if n mod 2 ==0 ,then 2 medians,first val[floor((n/2)-1)] and second val[floor(n/2)]
    print("median of the data frame is ",df_sorted.iloc[int((no_of_tuples/2)-1),1]," ",df_sorted.iloc[int(no_of_tuples/2),1])
```
